### Funciones para obtener datos

Debe conocer la estructura de los archivos para entender lo que debe retornar.

```python
def leerAntenas(nombre_archivo):
    # nombre_archivo: str
    # TODO
    #LISTO
    antenas = []
    archivo = open(nombre_archivo+'.txt','r')
    for linea in archivo:
        codigo = linea.strip().split(',')
        antenas.append((float(codigo[0]),float(codigo[1]),float(codigo[2])))
    archivo.close()
    return antenas
print leerAntenas('antenas2')
```
